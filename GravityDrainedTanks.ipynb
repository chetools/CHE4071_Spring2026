```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
g = 9.8 # m/s2
A = 1. #m2
Ah = 0.1 #m2
h0 = 0.3 #m3/s
```

```python
def rhs(t, h):
    qin = 2. #m3/s
    return (qin - Ah* np.sqrt(2*g)* h**0.5)
```

```python
tend = 300 #s
tplot = np.linspace(0,tend,200)
```

```python
res = solve_ivp(rhs, (0,tend), [h0], dense_output=True)
res
```

```python
hplot = res.sol(tplot)[0]
```
