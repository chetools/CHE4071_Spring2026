```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
alpha = Ah*np.sqrt(2*g)

def rhs(t, vec):
    h1, h2, eint = vec
    q1, q2 = alpha*np.sqrt(h1), alpha*np.sqrt(h2)

    e = hsp - h2
    qin = qin0 + kc*e + ki*eint
    return [(qin-q1)/A, (q1-q2)/A, e]
```

```python
g = 9.8 # m/s2
A = 1. #m2
Ah = 0.1 #m2
qin0 = 1. #m3/s
h0 = ((qin0/Ah)**2)/(2*g)
hsp = 1.2 * h0
kc =0.1 #Proportional Controller gain
ki = 0.03 #Integral Controller gain
tend = 1000.
tplot = np.linspace(0,tend,1000)
h0, hsp
```

```python
res = solve_ivp(rhs, (0, tend), [h0, h0, 0.],  method='Radau', dense_output=True)
h1plot, h2plot, eint_plot  = res.sol(tplot)
```

```python
fig=make_subplots()
fig.add_scatter(x=tplot, y=h1plot, mode='lines', name='h1')
fig.add_scatter(x=tplot, y=h2plot, mode='lines', name='h2')
fig.update_layout(width=600, height=400, template='plotly_dark')
```
