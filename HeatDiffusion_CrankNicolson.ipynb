```python
#r = alpha*dt/(2*dx**2)
r = 0.5
```

```python
alpha = 1e-4 #m2/s
dt = 0.1 #s
dx =  0.01 #m
r = alpha*dt/2/dx**2
r
```

```python
N = 50
T = np.full(N, 20.)
T[0]=40.
T[-1]=100.
```

```python
ab = np.vstack( (np.full(N-2, -r), np.full(N-2, 1+2*r), np.full(N-2, -r)))
```

```python
idx = np.arange(N)
fig = make_subplots()
fig.add_scatter(x=idx, y=T, mode='lines', name='t=0')
for n in range(10000):
    b = (1-2*r)*T[1:-1] + r*T[0:-2] + r*T[2:]
    b[0] += r*T[0]
    b[-1] += r*T[-1]
    T[1:-1] = sp.linalg.solve_banded((1,1), ab, b)
    if n % 20 == 0:
        fig.add_scatter(x=idx, y=T, mode='lines', name=f't={n+1}')
fig
```
