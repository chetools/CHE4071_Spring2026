```python
q=100. #L/min
Cain = 1. #mol/L
Tin = 350. #K
V = 100. #L
rho = 1e3 #g/L
C = 0.239 #J/(g K)
negHr = 5e4 #J/mol
ER = 8750. #K
k0 = 7.2e10 #1/min
UA = 5e4 #J/(min K)

Tc0 = 300. #K
Tc = 305. #K
Ca0 = 0.5 #mol/L
T0 = 350. #K
```

```python
def rhs(t, vec):
    Ca, T = vec
    k = k0*np.exp(-ER/T)
    dCa = q*(Cain-Ca)/V - k*Ca
    dT = q*(Tin - T)/V - UA*(T-Tc)/(rho*V*C) + negHr*k*Ca/(rho*C)
    return dCa, dT
```

```python
tend= 10.
tplot = np.linspace(0,tend,500)
res = sp.integrate.solve_ivp(rhs, (0,tend), (Ca0, T0), method='Radau', dense_output = True)
```

```python
Caplot, Tplot = res.sol(tplot)
```

```python
fig = make_subplots(rows=1, cols=2)
fig.add_scatter(x=tplot, y=Caplot, name='Ca', row=1, col=1)
fig.add_scatter(x=tplot, y=Tplot, name='T', row=1, col=2)
fig
```
