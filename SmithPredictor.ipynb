```python
Kc = 5.
taui = 10.
Gc = Kc*(1+1/(taui*s))
K = 1.
tau = 2.
theta = 10.5

K_tilde = 0.95
tau_tilde = 2.2
theta_tilde = 10.

Gp = K*exp(-theta*s)/(tau*s+1)
Gp_tilde = K_tilde/(tau_tilde*s+1)
t, y = sim(Gc*Gp/(1+Gc*Gp), lambda t: 1., N=1000, dt=0.1)
```

```python
fig=make_subplots()
fig.add_scatter(x=t,y=y, mode='lines')
fig.update_layout(width=600,height=400)
```

```python
M = Gc*Gp/(1+Gc*Gp_tilde*(1-exp(-theta_tilde*s)))
```

```python
t,y_servo = sim(M/(1+M), lambda t: 1, N=1000, dt=0.1 )
Gd = Gp
t,y_disturb = sim(Gd/(1+M), lambda t: t/10 if t<10 else 1., N=1000, dt=0.1  )
```

```python
fig=make_subplots()
fig.add_scatter(x=t,y=y_servo, mode='lines', name='servo')
fig.add_scatter(x=t,y=y_disturb, mode='lines', name='disturb reject')
fig.update_layout(width=600,height=400)
```

```python
#combination of setpoint change with ramp in disturbance
fig=make_subplots()
fig.add_scatter(x=t,y=y_servo+y_disturb, mode='lines')
fig.update_layout(width=600,height=400)
```
