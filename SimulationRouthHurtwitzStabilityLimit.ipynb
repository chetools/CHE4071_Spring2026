#P-control, 3 1st order in series

```python
K = 1.
Kc = 8./K + 0.1
tau = 1.5
```

```python
Gp = K/(tau*s + 1)**3
Gc = Kc
Gclosed = Gc * Gp/(1+Gc*Gp)
t,h=sim(Gclosed, lambda t: 1, N=5000, dt=0.1)
```

```python
fig=make_subplots()
fig.add_scatter(x=t, y=h, mode='lines')
fig.update_layout(width=600, height=400)
```

#PI control, 2 first order in series

```python
K = 1.
Kc = 2.
tau = 1.5
taui = Kc*K*tau/(2*(Kc*K+1))
```

```python
Gp = K/(tau*s + 1)**2
Gc = Kc*(1 + 1/(taui*s))
Gclosed = Gc * Gp/(1+Gc*Gp)
```

```python
t,h=sim(Gclosed, lambda t: 1, N=500, dt=0.1)
fig=make_subplots()
fig.add_scatter(x=t, y=h, mode='lines')
fig.update_layout(width=600, height=400)
```
