```python
#1 is slave; 2 is master
Kc1 = 1.
Kc2 = 1.
taui2 = 1.

Gp1 = 1.5/(2*s + 1)
Gc1 = Kc1  #slave controller is almost always just a proportional controller
Gd1 = 1/(2.5*s + 1)
Gp2 = 1./(20.*s + 1)
Gc2 = Kc2 *(1 + 1/(taui2*s))
```

```python
beta = Gc2*Gc1*Gp1*Gp2/(1+Gc1*Gp1)
alpha = Gd1*Gp2/(1+Gc1*Gp1)
```

```python
#inner loop disturbance
t,y_disturb = sim(alpha/(1+beta), lambda t: 1., N=500, dt=1.)
fig=make_subplots()
fig.add_scatter(x=t, y=y_disturb, mode='lines')
fig.update_layout(width=600, height=400)
fig.show()
```

```python
#outer setpoint change
t,y_setpoint = sim(beta/(1+beta), lambda t: 1., N=500, dt=1.)
fig=make_subplots()
fig.add_scatter(x=t, y=y_setpoint, mode='lines')
fig.update_layout(width=600, height=400)
fig.show()
```

```python
fig=make_subplots()
fig.add_scatter(x=t, y=y_setpoint+y_disturb, mode='lines')
fig.update_layout(width=600, height=400)
fig.show()
```
