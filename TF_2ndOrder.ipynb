```python
g=9.8 #m/s2
rho = 1000 #kg/m3
L = 1. #m
mu = 1e-3
pi = 3.14
D = 0.01 #1 cm
A = (pi*D**2)/4

alpha = 1/(rho * L)
beta = 8*pi*mu/(rho*A)
gamma = g/L

K = alpha/gamma
tau = 1/np.sqrt(gamma)
zeta = beta/(2*np.sqrt(gamma))
```

```python
K, tau, zeta
```

```python
Gp = K/(tau**2 * s**2 + 2*zeta*tau*s + 1)
t, h = sim(Gp, lambda t: 5e3, N=1000, dt=0.1)
fig = make_subplots()
fig.add_scatter(x=t, y=h)
```
