```python
wn = 1/tau #natural frequency rad/s
```

```python
Gp = K/(tau**2 * s**2 + 2*zeta*tau*s + 1)
```

```python
t, h = sim(Gp, lambda t: np.sin(0.2*wn*t), N=1000, dt=0.5)
```
