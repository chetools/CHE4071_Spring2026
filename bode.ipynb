```python
#first order system with time delay
Kc = 20
K = 1.
tau = 2.
theta = 0.1 #time delay
Gc = Kc
G = K*exp(-theta*s)/((tau*s+1))
fig, mag, angle = bode(Gc*G, w=np.logspace(-3,2.,1000))
```

```python
fig
```

```python
#phase angle for a time delay is -theta*w
```

```python
t,y = sim(Gc*G/(1+Gc*G), lambda t: 1., N=1000, dt=0.1)
fig=make_subplots()
fig.add_scatter(x=t, y=y, mode='lines')
```

```python
#Smith Predictor
Kc = 10
K = 1.
tau = 2.
theta = 0.1 #time delay
Gc = Kc
G = K*exp(-theta*s)/((tau*s+1))
Gtilde =  K/((tau*s+1))

#closed loop response with Smith Predictor
Gcl_SP = Gc*G/(1+Gc*Gtilde)
t,y = sim(Gcl_SP, lambda t: 1., N=1000, dt=0.1)
fig=make_subplots()
fig.add_scatter(x=t, y=y, mode='lines')
```
