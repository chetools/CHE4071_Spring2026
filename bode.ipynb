```python
#first order system with time delay
Kc = 10
K = 1.
tau = 2.
theta = 0.1 #time delay
G = Kc*K*exp(-theta*s)/((tau*s+1))
fig, mag, angle = bode(G, w=np.logspace(-3,2,1000))
```
