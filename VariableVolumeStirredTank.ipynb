```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
def rhs(t, vec):
    h, Ca = vec
    q = alpha*np.sqrt(h)
    Cain = Cain0 + 0.3*Cain0*np.sin(5*2*np.pi*t)

    dh = (qin - q)/A
    dCa = ((Cain*qin - Ca*q)/A - Ca*dh)/h


    return [dh, dCa]
```

```python
g = 9.8 # m/s2
A = 1. #m2
Ah = 0.1 #m2
alpha = Ah*np.sqrt(2*g)
qin0 = 1. #m3/s
h0 = ((qin0/Ah)**2)/(2*g)

Cain0 = 1. #mol/m3

qin = 1.2*qin0

tend = 100.
tplot = np.linspace(0,tend,1000)
```

```python
res = solve_ivp(rhs, (0, tend), [h0, Cain0],  method='Radau', dense_output=True, rtol=1e-8, atol=1e-8)
hplot, Caplot  = res.sol(tplot)
```

```python
fig=make_subplots(rows=1, cols=2)
fig.add_scatter(x=tplot, y=hplot, mode='lines', name='h', row=1, col=1)
fig.add_scatter(x=tplot, y=Caplot, mode='lines', name='Ca', row=1, col=2)
fig.update_layout(width=800, height=400, template='plotly_dark')
```
